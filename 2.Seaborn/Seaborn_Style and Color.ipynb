```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('poster', font_scale=2)
sns.countplot(x='sex', data=tips)
```
